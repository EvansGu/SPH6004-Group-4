```python
import re

# Extract 'HISTORY', 'INDICATION', and 'IMPRESSION'

def extract_notes(text):
    pattern_history = r'HISTORY:([^:]+)'
    pattern_indication = r'INDICATION:([^:]+)'
    pattern_impression = r'IMPRESSION:([^:]+)'

    history_match = re.search(pattern_history, text, re.IGNORECASE)
    history = history_match.group(1).strip() if history_match else ""

    indication_match = re.search(pattern_indication, text, re.IGNORECASE)
    indication = indication_match.group(1).strip() if indication_match else ""

    impression_match = re.search(pattern_impression, text, re.IGNORECASE)
    impression = impression_match.group(1).strip() if impression_match else ""

    notes = 'History: ' + history + '\nIndication: ' + indication + '\nImpression: ' + impression

    words_to_remove = ["TECHNIQUE", "COMPARISON", "FINDINGS"]

    for word in words_to_remove:
      notes = notes.replace(word, "")

    return notes
```

```python
notes['text'] = notes['text'].apply(extract_notes)
```

```python
import scispacy
import spacy

import en_core_sci_sm
import en_core_sci_md
import en_ner_bc5cdr_md

from spacy import displacy
```

```python
def extract_disease_entities(text, model):
    nlp_model = spacy.load(model)
    doc = nlp_model(text)
    disease_entities = [ent.text for ent in doc.ents if ent.label_ == 'DISEASE']
    return disease_entities

def extract_all_disease_entities(text, models):
    all_disease_entities = []
    for model in models:
        disease_entities = extract_disease_entities(text, model)
        all_disease_entities.extend(disease_entities)
    return ', '.join(all_disease_entities)

# Define the models to use
models = ["en_core_sci_sm", "en_core_sci_md", "en_ner_bc5cdr_md"]

# Apply the function to each row of the DataFrame
notes['disease'] = notes['text'].apply(lambda x: extract_all_disease_entities(x, models))
```

```python
# Split the list of diseases into individual strings
disease_lists = notes['disease'].str.split(',')

# Get unique list of diseases
L = set()
for sublist in disease_lists:
    for disease in sublist:
        L.add(disease)

# Create binary indicators for each disease
for disease in L:
    notes[disease.strip()] = notes['disease'].apply(lambda x: 1 if disease.strip() in x else 0)

# Drop the original 'disease' column if needed
notes = notes.drop(columns=['text', 'disease'])
```
